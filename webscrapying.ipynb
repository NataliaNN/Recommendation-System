```
url_profile = 'http://api.steampowered.com/ISteamUser/GetPlayerSummaries/v0002/?key=6DB01CC1FBADFFAAA11BD70464928248&steamids={}'
#s1 = url_profile.format(ID[0])
#Summary = json.loads(requests.get(s1).text)
#Summary
```

```
url_profile = 'http://api.steampowered.com/ISteamUser/GetPlayerSummaries/v0002/?key=6DB01CC1FBADFFAAA11BD70464928248&steamids={}'
profile = []

for i in range(2):
  url = url_profile.format(ID[i])
  try:
    content = json.loads(requests.get(url).text)['response']['players'][0]
    content = {k:v for k, v in content.items() if k in ['lastlogoff','timecreated','loccountrycode']}
    profile.append(content)
  except:
    profile.append(np.nan)
```
